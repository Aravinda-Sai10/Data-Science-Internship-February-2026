```python
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
```

## Task 1 – Replace Hardcoded Prompt

### Goal

Convert a hardcoded prompt into a reusable PromptTemplate.

### Approach

Instead of using string formatting, we use LangChain PromptTemplate to make prompts reusable.

```python
template = PromptTemplate(

input_variables=["topic"],

template="Explain {topic} in simple terms for beginners"

)

result = template.format(topic="Artificial Intelligence")

print(result)
```

## Task 2 – Multi Input Prompt System

### Goal

Create a prompt template that accepts multiple inputs such as topic, audience and tone.

```python
multi_prompt = PromptTemplate(

input_variables=["topic","audience","tone"],

template="Explain {topic} for {audience} in a {tone} tone"

)

print(multi_prompt.format(

topic="AI",
audience="beginners",
tone="friendly"

))

print(multi_prompt.format(

topic="Python",
audience="kids",
tone="fun"

))

print(multi_prompt.format(

topic="Deep Learning",
audience="engineers",
tone="technical"

))
```

## Task 3 – Prompt Variations Engine

### Goal

Create different prompt templates for teaching, interview and storytelling styles.

```python
teaching_prompt = PromptTemplate(

input_variables=["topic"],

template="Explain {topic} clearly step by step"

)

interview_prompt = PromptTemplate(

input_variables=["topic"],

template="Ask 3 interview questions about {topic}"

)

story_prompt = PromptTemplate(

input_variables=["topic"],

template="Explain {topic} as a story"

)

topic="Machine Learning"

print(teaching_prompt.format(topic=topic))

print(interview_prompt.format(topic=topic))

print(story_prompt.format(topic=topic))
```

## Task 4 – ChatPromptTemplate System

### Goal

Build a chat based prompt system using roles like teacher, interviewer and motivator.

```python
chat_template = ChatPromptTemplate.from_template(

"Act as a {role} and explain {topic}"

)

print(chat_template.format(

role="teacher",
topic="Neural Networks"

))

print(chat_template.format(

role="interviewer",
topic="Neural Networks"

))

print(chat_template.format(

role="motivator",
topic="Neural Networks"

))
```

## Task 5 – Input Validation Layer

### Goal

Validate inputs before prompt generation.

```python
def validate_inputs(audience,tone):

    valid_audience=["beginner","intermediate","expert"]

    valid_tone=["formal","casual","fun"]

    if audience not in valid_audience:

        print("Invalid audience detected → default applied")

        audience="beginner"

    if tone not in valid_tone:

        print("Invalid tone detected → default applied")

        tone="formal"

    return audience,tone


a,t = validate_inputs("beginner","fun")

print(a,t)
```

## Task 6 – Prompt Generator App

### Goal

Create a reusable function that generates prompts dynamically.

```python
generator = PromptTemplate(

input_variables=["topic","audience","tone","style"],

template="Explain {topic} for {audience} in a {tone} {style} style"

)

def generate_prompt(topic,audience,tone,style):

    audience,tone = validate_inputs(audience,tone)

    prompt = generator.format(

    topic=topic,
    audience=audience,
    tone=tone,
    style=style

    )

    return prompt


print(generate_prompt(

"Neural Networks",
"beginner",
"fun",
"storytelling"

))
```

## Task 7 – Template Reusability Test

### Goal

Test one template with multiple inputs to prove reusability.

```python
reuse_template = PromptTemplate(

input_variables=["topic"],

template="Explain fundamentals of {topic}"

)

topics=[

"Artificial Intelligence",
"Machine Learning",
"Deep Learning",
"NLP",
"Computer Vision"

]

for topic in topics:

    print(reuse_template.format(topic=topic))
```
