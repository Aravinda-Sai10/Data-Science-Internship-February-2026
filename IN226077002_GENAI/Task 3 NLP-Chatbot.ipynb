```python
# importing required modules from transformers
# AutoTokenizer → converts text to tokens
# AutoModelForSeq2SeqLM → loads transformer model for text generation
transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
```

```python
model_name="google/flan-t5-base" # selecting FLAN-T5 pretrained transformer model from HuggingFace

tokenizer=AutoTokenizer.from_pretrained(model_name) # loading tokenizer for text processing

model=AutoModelForSeq2SeqLM.from_pretrained(model_name) # loading pretrained transformer model

print("Model Loaded Successfully")
```

```python
# starting chatbot interface
print("Chatbot: Hello! I am your AI assistant. Type exit to stop.")

# continuous conversation loop
while True:

    user_input=input("User: ")

    if user_input.lower() in ["exit","quit"]:

        print("Chatbot: Goodbye!")
        break

    prompt="Answer this question clearly: "+user_input

    inputs=tokenizer(prompt,return_tensors="pt")

    outputs=model.generate(

        inputs.input_ids,

        max_length=100,

        temperature=0.7

    )

    response=tokenizer.decode(

        outputs[0],

        skip_special_tokens=True

    )

    print("Chatbot:",response) # printing chatbot response
```

## Code Explanation

This implementation uses FLAN-T5 transformer model from HuggingFace.

Steps followed:

1 Load pretrained transformer model  
2 Accept user input  
3 Convert input into tokens  
4 Generate response using transformer  
5 Decode response  
6 Continue conversation until exit  

The chatbot uses prompt engineering to improve response quality.
