# Task 1 : Conceptual Understanding

## 1 What is difference between Love and love in NLP?

In NLP, Love and love are treated as different tokens because NLP systems are case sensitive. Computers treat uppercase and lowercase differently.

Example:

Love → Token 1  
love → Token 2  

To avoid this issue we convert text into lowercase. This process is called case normalization.

---

## 2 What happens if stopwords are not removed?

If stopwords are not removed:

• Noise increases  
• Model training becomes slow  
• Important words lose importance  
• Accuracy may reduce  

Example:

Sentence:
I am going to the market

Without removal:
[I, am, going, to, the, market]

After removal:
[going, market]

---

## 3 Two real world cases where stopword removal is harmful

Case 1 : Sentiment Analysis

Sentence:
I am not happy

If we remove "not":
happy

Meaning becomes opposite.

Case 2 : Question answering systems

Example:
To be or not to be

Removing stopwords destroys meaning.

---

## 4 Difference between Stemming and Lemmatization

Stemming:
Cuts words roughly  
Example:
studies → studi

Lemmatization:
Uses vocabulary and produces meaningful words

Example:
studies → study

Stemming is faster but less accurate.
Lemmatization is slower but more accurate.

# Task 2 : Advanced Preprocessing Function

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):

    if not isinstance(text,str):
        return [],""

    if text.strip()=="":
        return [],""

    # remove urls
    text=re.sub(r'http\S+|www\S+','',text)

    # remove emails
    text=re.sub(r'\S+@\S+','',text)

    # remove numbers
    text=re.sub(r'\d+','',text)

    # remove special characters and emojis
    text=re.sub(r'[^\w\s]',' ',text)

    # convert lowercase
    text=text.lower()

    # remove repeated characters
    text=re.sub(r'(.)\1{2,}',r'\1',text)

    # remove extra spaces
    text=re.sub(r'\s+',' ',text).strip()

    tokens=text.split()

    clean_tokens=[]

    for word in tokens:

        if len(word)<=2 and word not in ['no','not']:
            continue

        clean_tokens.append(word)

    clean_sentence=" ".join(clean_tokens)

    return clean_tokens,clean_sentence
```

# Task 3 : Stress Testing

```python
test_sentences=[

"Get 100% FREE access now!!!",

"I absolutely looooved this product 😍😍",

"Worst service ever... 0/10",

"Call me at 9876543210",

"This is THE best course!!!",

"Visit https://openai.com now!",

"Nooooo this is baaad!!!",

"OK OK OK I got it",

"Win $$$ now!!! Limited offer!!!",

"I am not happy with this"

]
```

```python
for sentence in test_sentences:

    tokens,clean=preprocess_text(sentence)

    print("Original Text :",sentence)

    print("Cleaned Tokens :",tokens)

    print("Cleaned Sentence :",clean)

    print()
```

# Task 4 : Token Analytics

```python
for sentence in test_sentences:

    tokens,clean=preprocess_text(sentence)

    total_tokens=len(tokens)

    unique_tokens=len(set(tokens))

    if total_tokens>0:
        avg_length=sum(len(word) for word in tokens)/total_tokens
    else:
        avg_length=0

    print("Sentence :",sentence)

    print("Total tokens :",total_tokens)

    print("Unique tokens :",unique_tokens)

    print("Average token length :",round(avg_length,2))

    print()
```

## Analysis Answers

### Which sentence had most noise?

Win $$$ now Limited offer

Because it contains symbols, promotional noise and repeated punctuation.

---

### Which sentence retained most meaningful tokens?

I absolutely loved this product

Because after cleaning it still preserves the main meaning.

# Task 5 : Frequency Analysis

```python
all_tokens=[]

for sentence in test_sentences:

    tokens,clean=preprocess_text(sentence)

    all_tokens.extend(tokens)

freq=Counter(all_tokens)

print("Top 10 frequent words")

print(freq.most_common(10))

print()

print("Top 5 least frequent words")

print(freq.most_common()[-5:])
```

# Task 6 : Full Pipeline

```python
def full_pipeline(text_list):

    all_tokens=[]

    clean_sentences=[]

    for text in text_list:

        tokens,clean=preprocess_text(text)

        all_tokens.extend(tokens)

        clean_sentences.append(clean)

    return {

        "tokens":all_tokens,

        "clean_sentences":clean_sentences

    }

result=full_pipeline(test_sentences)

print(result)
```
