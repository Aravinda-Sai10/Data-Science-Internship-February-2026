# 1. Unique Words in a Sentence

```python
sentence = input("Enter sentence: ") # taking input from user

# converting sentence to lowercase
sentence = sentence.lower()

# splitting sentence into words
words = sentence.split()

# removing duplicates using set
unique_words = set(words)

# printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# 2. Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# assume first employee as highest
highest_name = ""
highest_salary = 0

# checking each employee
for name in employees:
    salary = employees[name]
    
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name

print("Highest Salary:", highest_name, "-", highest_salary)
```

# 3. Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 10, 66]

# assume first element as max and min
max_val = numbers[0]
min_val = numbers[0]

# checking each number
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

# printing results
print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

# 4. Count Products Above a Price Threshold


```python
prices = [450, 1200, 899, 1500, 300]

count = 0

# checking each price
for price in prices:
    if price > 1000:
        count = count + 1

print("Products above 1000:", count)
```

# 5. Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

# counting present days
for day in attendance:
    if day == "P":
        present_count = present_count + 1

# total days
total_days = len(attendance)

# calculating percentage
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

# 7. Count Character Frequency


```python
text = "pythonn"

char_count = {}

# checking each character
for ch in text:
    if ch in char_count:
        char_count[ch] = char_count[ch] + 1
    else:
        char_count[ch] = 1

print(char_count)
```

# 9. Check if a Key Exists in a Dictionary



```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key = "Ravi"   # key to check

# checking if key exists
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

# 10. Calculate Average Marks


```python
marks = [60, 70, 80, 80]   # marks

total = 0

# finding total marks
for m in marks:
    total = total + m

# calculating average
average = total / len(marks)

# result
print("Average Marks:", average)
```
