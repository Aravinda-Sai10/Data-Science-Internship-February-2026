# 1. Social Media – Post Engagement Analyzer

```python
def analyze_post(likes_list):
    total_likes = 0
    
    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like
    
    # Decide post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    return total_likes, status


# Sample likes data
likes = [200, 150, 300, 250, 150]

total, result = analyze_post(likes)

print("Total Likes:", total)
print("Post Status:", result)
```

# 3. Agriculture – Rainfall Adequacy Checker


```python
def analyze_rainfall(rain_data, required_level):
    total = 0
    
    # Calculate total rainfall
    for r in rain_data:
        total += r
    
    avg = total / len(rain_data)
    
    # Check rainfall status
    if avg >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    return avg, status


# Sample rainfall data
rainfall = [60, 80, 75, 70, 75]
required = 70

average, result = analyze_rainfall(rainfall, required)

print("Average Rainfall:", round(average))
print("Rainfall Status:", result)
```

# 4. Social Media – Duplicate Account Detection

```python
def check_duplicates(usernames):
    unique_users = set(usernames)
    
    # Compare lengths
    if len(unique_users) != len(usernames):
        return "Yes"
    else:
        return "No"


# Sample usernames list
users = ["john123", "alice", "rahul", "john123", "sneha"]

result = check_duplicates(users)

print("Duplicate Accounts Found:", result)
```

# 6. Agriculture – Premium Crop Price Filter


```python
def filter_premium(prices):
    premium = []
    
    # Check each price
    for p in prices:
        if p > 2000:
            premium.append(p)
    
    return premium


# Sample crop prices
crop_prices = [1500, 2500, 1800, 3200, 1900]

premium_list = filter_premium(crop_prices)

print("Premium Crops:", premium_list)
```

# 7. System Monitoring – Application Health Checker


```python
def check_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    return errors, status


# Sample error count
error_count = 7

count, result = check_health(error_count)

print("Error Count:", count)
print("System Status:", result)
```

# 8. Banking – Daily Transaction Limit Checker

```python
def check_transaction(amount):
    limit = 50000
    
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    return amount, status


# Sample transaction amount
txn_amount = 60000

amt, result = check_transaction(txn_amount)

print("Transaction Amount:", amt)
print("Transaction Status:", result)
```

# 9. E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance):
    present = 0
    
    # Count present days
    for day in attendance:
        if day == "P":
            present += 1
    
    percentage = (present / len(attendance)) * 100
    
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    return percentage, status


# Sample attendance data
attendance_list = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]

percent, result = check_exam_eligibility(attendance_list)

print("Attendance Percentage:", round(percent, 1))
print("Exam Eligibility:", result)
```

# 10. Smart Electricity Bill Analyzer


```python
def calculate_bill(units):
    bill = 0
    
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status


# Sample units consumed
units_used = 250

total_bill, usage = calculate_bill(units_used)

print("Units Consumed:", units_used)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage)
```
