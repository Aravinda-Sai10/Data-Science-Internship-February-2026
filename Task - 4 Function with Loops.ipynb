# 1. Mobile Recharge Validation System


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599] # List of valid recharge plans
    
    # Check minimum recharge amount
    if amount < 50:
        print("Recharge amount should be at least ₹50.")
        return False
    
    # Check if amount is in valid plans
    if amount not in valid_plans:
        print("Invalid recharge plan selected.")
        print("Available plans are:", valid_plans)
        return False
    
    # If all conditions are satisfied
    return True

print("=== Mobile Recharge System ===")
while True:
    try:
        # Taking input from user
        user_input = int(input("Enter recharge amount: "))
        
        # Call validation function
        if validate_recharge(user_input):
            print("Recharge successful! Thank you.")
            break   # Exit loop if valid
        else:
            print("Please try again.\n")
    
    except ValueError:
        # Handling wrong input like text instead of numbers
        print("Invalid input! Please enter numbers only.\n")
```

# 2. Inventory Reorder Alert System


```python
def check_stock(products):
    for item, quantity in products.items():
        
        # Check if stock is less than 15
        if quantity < 15:
            print(item, "Reorder Alert (Stock:", quantity, ")")
        else:
            print(item, "Stock OK (Stock:", quantity, ")")


# Sample product data
product_stock = {
    "Rice": 40,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 8,
    "Wheat": 15
}

print("=== Warehouse Stock Status ===")

# Calling the function
check_stock(product_stock)
```

 # 3. Student Result Processing System


```python
# Function to calculate average and result
def check_result(marks):
    total = 0
    # Calculate total using loop
    for m in marks:
        total += m
    
    avg = total / len(marks)
    
    # Return result based on average
    if avg >= 50:
        return avg, "Pass"
    else:
        return avg, "Fail"


# Sample marks list
student_marks = [60, 45, 70, 55, 40]

average, result = check_result(student_marks)

print("Average Marks:", average)
print("Result:", result)
```

# 4. Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12
    
    fare = base_fare + (distance * per_km)
    
    # Add peak hour charges
    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)
    
    return fare


print("=== Cab Fare Calculator ===")

while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_time = input("Is it peak hour? (yes/no): ")
        
        total_fare = calculate_fare(km, peak_time)
        print("Total Fare: ₹", round(total_fare, 2))
        
        # Retry option
        again = input("Do you want to calculate again? (yes/no): ")
        if again.lower() != "yes":
            print("Thank you for using the cab service!")
            break
    
    except ValueError:
        print("Invalid input! Please enter correct values.\n")
```

# 5. Employee Attendance Eligibility Checker


```python
def check_attendance(att_list):
    present = 0
    
    # Count number of presents
    for day in att_list:
        if day == "P":
            present += 1
    
    percentage = (present / len(att_list)) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Sample attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

percent, status = check_attendance(attendance)

print("Attendance Percentage:", round(percent, 2), "%")
print("Status:", status)
```

# 6. Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True
    
    # Apply rules
    if len(password) < 8:
        return "Weak Password (Too short)"
    elif not has_digit:
        return "Weak Password (No digit)"
    elif not has_special:
        return "Weak Password (No special character)"
    else:
        return "Strong Password"


# Sample input
pwd = input("Enter password: ")

result = check_password(pwd)
print("Result:", result)
```
