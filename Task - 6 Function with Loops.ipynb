# 1. Smart Parking Lot Management System


```python
def manage_parking(logs, capacity):
    parked = 0
    peak = 0
    
    for entry in logs:
        if entry == "IN":
            parked += 1
        elif entry == "OUT" and parked > 0:
            parked -= 1
        
        # Track peak usage
        if parked > peak:
            peak = parked
    
    # Checking parking status
    if parked > capacity:
        status = "Over Capacity Alert"
    else:
        status = "Available"
    
    return parked, peak, status


capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

current, peak_usage, parking_status = manage_parking(vehicle_logs, capacity)

print("Currently Parked Vehicles:", current)
print("Peak Parking Usage:", peak_usage)
print("Parking Status:", parking_status)
```

# 2. Online Food Delivery Time Estimator


```python
def estimate_delivery(distance, traffic, weather):
    time = distance * 5
    
    # Traffic adjustment
    if traffic.lower() == "high":
        time += 15
    elif traffic.lower() == "medium":
        time += 10
    
    # Weather adjustment
    if weather.lower() == "rainy":
        time += 10
    elif weather.lower() == "storm":
        time += 20
    
    return time

dist = 8
traffic_level = "High"
weather_cond = "Rainy"
eta = estimate_delivery(dist, traffic_level, weather_cond)
print("Estimated Delivery Time:", eta, "minutes")
```

# 3. Movie Theatre Seat Occupancy Analyzer


```python
def analyze_occupancy(total_seats, bookings):
    booked_count = len(bookings)
    
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    
    # show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    
    return occupancy, status


# Sample input
total = 200
booked_seats = [1] * 150   # 150 bookings

occ, result = analyze_occupancy(total, booked_seats)

print("Occupancy:", round(occ), "%")
print("Show Status:", result)
```

# 4. Cloud Server Load Classification System


```python
def classify_load(cpu_readings):
    total = 0
    
    # Calculate total CPU usage
    for val in cpu_readings:
        total += val
    
    avg = total / len(cpu_readings)
    
    # Classification
    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    return avg, status


# Sample input
cpu_data = [45, 60, 70, 85, 90]

average, result = classify_load(cpu_data)

print("Average CPU Load:", round(average), "%")
print("Server Status:", result)
```

# 5. Smart Classroom Resource Usage Monitor


```python
def monitor_usage(resources):
    overused = []
    
    for name, hours in resources.items():
        if hours > 8:   # threshold for overuse
            overused.append(name)
    
    # Energy alert check
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"
    
    return overused, alert


# Sample input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused_resources, energy_alert = monitor_usage(usage_data)

print("Overused Resources:", ", ".join(overused_resources))
print("Energy Alert:", energy_alert)
```

# 6. Online Event Registration Capacity Controller


```python
def manage_registrations(capacity, total_reg):
    
    if total_reg <= capacity:
        confirmed = total_reg
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = total_reg - capacity
        status = "Closed"
    
    return confirmed, waitlisted, status


# Sample input
event_capacity = 100
registrations = 105

confirmed_users, waitlist_users, reg_status = manage_registrations(event_capacity, registrations)

print("Confirmed Registrations:", confirmed_users)
print("Waitlisted Users:", waitlist_users)
print("Registration Status:", reg_status)
```
